```python
import numpy as np
import scanpy as sc
import seaborn as sns
import os
import pandas as pd
import matplotlib.pyplot as plt
import warnings
import sys
import pickle as pkl
path_helper = ["C:\\","Users","vfriedrich","projects","monkey_IZI","git_documentation","scRNAseq_cross_species_primate_human","analysis","helper"]
sys.path.append(os.path.join(*path_helper))
import helperVDF as h
#import decoupler
print(sys.executable)
import gseapy
from gseapy.plot import barplot, dotplot
warnings.filterwarnings("ignore")
import re
```

```python
pre = "MH114"
drive = 'F'
base_model_path,base_table_path,base_plots_path,base_anndata_objects = h.return_local_paths(drive = drive,
                                                                                            pre = pre,
                                                                                            add_path = True)
def remove_go(term):
    # Use regular expression to find and replace GO terms, ensuring the string is treated as a raw string
    new_term = re.sub(r'\s*\(GO:\d+\)', '', term)  # Corrected regex finds the GO term pattern and removes it
    return new_term.strip() 

col_human =  "#e35e28"
col_cyno ="#31c7ba"
_,base_table_path_MH113,_,_ = h.return_local_paths(drive = drive,pre = 'MH113',add_path = True)
```

## CD8 T - cluster 10

```python
celltype = 'CD8 T'
cluster_of_interest = '10'
```

```python
filtered_df_human = pd.read_csv(
    (os.path.join(base_table_path_MH113,'MH113_filtered_df_pathways_' + str(celltype) + '_' + str(cluster_of_interest) + '_' + 'human' + '.csv')),
index_col=0)

filtered_df_human['species'] = 'human'

filtered_df_cyno = pd.read_csv(
    (os.path.join(base_table_path_MH113,'MH113_filtered_df_pathways_' + str(celltype) + '_' + str(cluster_of_interest) + '_' + 'cyno' + '.csv')),
index_col=0)

filtered_df_cyno['species'] = 'cyno'

top_pathways = list(set(filtered_df_human.head(6).Term.values).union(set(filtered_df_cyno.head(6).Term.values)))

merged_df = pd.merge(filtered_df_human, filtered_df_cyno, on='Term', how='outer', suffixes=('_human', '_cyno'))

merged_df.to_csv(os.path.join(base_table_path,pre + '_merged_df_' + str(celltype) + '_' + str(cluster_of_interest) + '.csv'))

merged_df_topP = merged_df[merged_df['Term'].isin(top_pathways)]
merged_df_topP['Log Odds Ratio_cyno'].fillna(0, inplace=True)
merged_df_topP['Log Odds Ratio_human'].fillna(0, inplace=True)

merged_df_topP['short_term'] = merged_df_topP['Term'].apply(remove_go)

bar_width = 0.45
index = np.arange(len(merged_df_topP))
fig, ax = plt.subplots(figsize=(10, 6))
bars1 = ax.barh(index - bar_width/2, merged_df_topP['Log Odds Ratio_human'], bar_width, label='human', color=col_human)
bars2 = ax.barh(index + bar_width/2, merged_df_topP['Log Odds Ratio_cyno'], bar_width , label='cyno', color=col_cyno)




ax.set_xlabel('Log Odds Ratio')
ax.set_yticks(index)
ax.set_yticklabels(merged_df_topP['short_term'], fontweight='bold')
#ax.set_title(f'{species} - {celltype} - {cluster_of_interest}')
ax.invert_yaxis()  # Inverts the y-axis so the largest values are on top
ax.legend(loc='upper left', bbox_to_anchor=(1,1))

# Grid, layout, and save
plt.grid(True, linestyle='--', alpha=0.2)
plt.tight_layout()
plt.title(str(celltype) + ' - cluster ' + str(cluster_of_interest))
plt.savefig(os.path.join(base_plots_path, pre + '_' + str(celltype) + '_' + str(cluster_of_interest) + '_pathways.pdf'), dpi=300, bbox_inches='tight')
plt.show()
```
