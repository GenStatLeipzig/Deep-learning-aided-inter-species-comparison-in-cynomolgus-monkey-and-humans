```python
base_anndata_objects_MH51_c = os.path.join('F:\\monkey_IZI\\analysis','cluster_runs',
                                    'workspace-runVAE','output','MH51_c','anndata_objects')

adata_train_all_ct = sc.read_h5ad(os.path.join(base_anndata_objects_MH51_c,'MH51_c_adata_train.h5ad'))
adata_test_all_ct = sc.read_h5ad(os.path.join(base_anndata_objects_MH51_c,'MH51_c_adata_test.h5ad'))

adata_train_all_ct.obs.rename(columns={'species_x': 'species'}, inplace=True)
adata_test_all_ct.obs.rename(columns={'species_x': 'species'}, inplace=True)

adata_train_all_ct.obs['cluster_azimut1_5_scanvi_v2'] = adata_train_all_ct.obs['cluster_azimut1_5_scanvi'].replace(
    {'NK': 'NK+Proliferating', 'NK Proliferating': 'NK+Proliferating'})

adata_test_all_ct.obs['cluster_azimut1_5_scanvi_v2'] = adata_test_all_ct.obs['cluster_azimut1_5_scanvi'].replace(
    {'NK': 'NK+Proliferating', 'NK Proliferating': 'NK+Proliferating'})
```

```python
model_path = os.path.join(*["F:\\","monkey_IZI","analysis","cluster_runs","workspace-runVAE",
                            "output","MH51_c","models"])
model_dir_names = os.listdir(model_path)

warnings.filterwarnings("ignore")
model_dict = {}
for model_dir_name in model_dir_names:
    load_string = os.path.join(model_path,model_dir_name,'MH51_c_scgen_model_' + str(model_dir_name)+'.pt')
    model = scgen.SCGEN.load(load_string)
    model_dict[model_dir_name] = model
```

### Spearman R comparison

```python
df_spearman_06hr_human = pd.DataFrame(columns = ['sp_00hr','sp_06hr_pred','celltype'])
for celltype in celltypes:
    GE_06hr_pred_human = dict_mean_exp_dict_human[celltype]['mean_exp_decoded_adata_test_06hr_human_pred']
    GE_00hr_human = dict_mean_exp_dict_human[celltype]['mean_exp_adata_test_00hr_human']
    GE_06hr_human = dict_mean_exp_dict_human[celltype]['mean_exp_adata_test_06hr_human']
    df_base_06hr_human = pd.DataFrame([GE_00hr_human,GE_06hr_human,GE_06hr_pred_human],index= ['GE_00hr_human','GE_06hr_human','GE_06hr_pred_human']).T
    df_base_06hr_human = df_base_06hr_human[df_base_06hr_human['GE_06hr_human'] > 0]
    sp_00hr = spearmanr(df_base_06hr_human['GE_00hr_human'],df_base_06hr_human['GE_06hr_human'])[0]
    sp_06hr_pred = spearmanr(df_base_06hr_human['GE_06hr_pred_human'],df_base_06hr_human['GE_06hr_human'])[0]

    df_temp_00hr_human =  pd.DataFrame(df_base_06hr_human['GE_00hr_human']).rank(method='dense', ascending=False).astype(int)
    df_temp_00hr_human.columns = ['rank_GE_00hr']

    df_temp_06hr_human =  pd.DataFrame(df_base_06hr_human['GE_06hr_human']).rank(method='dense', ascending=False).astype(int)
    df_temp_06hr_human.columns = ['rank_GE_06hr']

    df_temp_06hr_pred_human =  pd.DataFrame(df_base_06hr_human['GE_06hr_pred_human']).rank(method='dense', ascending=False).astype(int)
    df_temp_06hr_pred_human.columns = ['rank_GE_06hr_pred']

    df_06hr_rank_comp_human = df_temp_06hr_pred_human.T.append(df_temp_06hr_human.T).append(df_temp_00hr_human.T).T

    df_06hr_rank_comp_human['celltype'] = celltype

    df_06hr_rank_comp_human.to_csv(os.path.join(base_table_path,pre + '_' + str(celltype) + '_df_06hr_rank_comp_human.csv'))
    
    df_spearman_06hr_human = df_spearman_06hr_human.append(
         pd.DataFrame([sp_00hr,sp_06hr_pred,celltype],
                      index = ['sp_00hr','sp_06hr_pred','celltype']).T)
```

## Plot scatter

```python
def make_scatter_rank_pp(species,timepoint,filtered=False,show_plot = True,
                         celltypes=celltypes,base_table_path=base_table_path,base_plots_path=base_plots_path,save =True,pre=pre):
    fig, axes = plt.subplots(len(celltypes), 2, figsize=(6, 2.5 * len(celltypes)))  


    for i, celltype in enumerate(celltypes):
        if filtered == False:
            df_temp_rank = pd.read_csv(
                os.path.join(base_table_path,
                             pre + '_' + str(celltype) + '_df_' + str(timepoint) +'_rank_comp_' + str(species) + '.csv'),
                index_col=0)
        else:
            df_temp_rank = pd.read_csv(
                os.path.join(base_table_path,
                             pre + '_' + str(celltype) + '_df_' + str(timepoint) +'_rank_comp_' + str(species) + '_rank_filtered.csv'),
                index_col=0)
        axes[i, 0].scatter(df_temp_rank['rank_GE_'+ str(timepoint)],df_temp_rank['rank_GE_00hr'],s=6,color = "#022a2a",alpha =0.9)  # Replace x1, y1 with data for the first plot
        axes[i, 0].set_title(species + ' - ' + celltype,fontsize =10)
        axes[i, 0].set_xlabel('rank '+ str(timepoint))
        axes[i, 0].set_ylabel('rank '+ str('00hr'))


        axes[i, 1].scatter(df_temp_rank['rank_GE_'+ str(timepoint)], df_temp_rank['rank_GE_'+ str(timepoint)+str('_pred')],s=6,color = "#022a2a",alpha =0.9)  # Replace x2, y2 with data for the second plot
        axes[i, 1].set_title(species + ' - ' + celltype,fontsize =10)
        axes[i, 1].set_xlabel('rank '+ str(timepoint))
        axes[i, 1].set_ylabel('rank '+ str(timepoint)+str('_pred'))

    plt.tight_layout()
    if save:
        if filtered == False:
            plt.savefig(os.path.join(base_plots_path,pre + '_scatter_all_ct_' + str(species) + '_'+ str(timepoint)+ '.pdf'),bbox_inches='tight')
        else:
            plt.savefig(os.path.join(base_plots_path,pre + '_scatter_all_ct_' + str(species) + '_'+ str(timepoint)+ '_rank_filtered.pdf'),bbox_inches='tight')
    if show_plot:
        plt.show()    
```

```python
for species in ['human','cyno']:
    for timepoint in ['06hr','24hr']:
        make_scatter_rank_pp(species=species,timepoint=timepoint,filtered=False,show_plot=True)
        #make_scatter_rank_pp(species=species,timepoint=timepoint,filtered=False,show_plot=True)
```
